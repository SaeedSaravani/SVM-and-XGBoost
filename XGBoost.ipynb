```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('training.csv')
df1 = pd.read_csv('testing.csv')
df.tail(17)
```

```python
X_train = df.drop('class', axis=1).to_numpy()
y_train = df['class'].to_numpy()
X_test = df1.drop('class', axis=1).to_numpy()
y_test = df1['class'].to_numpy()
```

```python
n_estimators_values = np.arange(5,20)
max_depth_values = np.arange(1,10)
max_feature_values = np.arange(1,28)

accs_poly = np.zeros((n_estimators_values.shape[0], max_depth_values.shape[0],max_feature_values.shape[0]))

for i, r in enumerate(n_estimators_values):
    for j, d in enumerate(max_depth_values):
        for k,f in enumerate(max_feature_values):
            RFC = RandomForestClassifier(n_estimators=r , max_depth=d,  max_features=f,random_state=None).fit(X_train, y_train)
            predictions = RFC.predict(X_test)
            accs_poly[i,j,k] = accuracy_score(y_test, predictions)


max_acc =np.amax(accs_poly)
print(np.where(max_acc==accs_poly))
print(max_acc)
```

```python
print(f"Best parameter: \nn_estimators:{n_estimators_values[7]:.2f}, max_depth: {max_depth_values[4]:.2f} ,max_feature: {max_feature_values[10]:.2f}")
```

```python
#accuracy on training set
RFC = RandomForestClassifier(n_estimators=12 , max_depth=5,  max_features=11,random_state=None).fit(X_train, y_train)
predictions = RFC.predict(X_train)
accs_poly = accuracy_score(y_train, predictions)
print(accs_poly)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(X_train, y_train).predict(X_test)
accs_poly = accuracy_score(y_test, y_pred)
print(accs_poly)
```

```python
from xgboost import XGBRFClassifier,XGBClassifier 


xgbrf = XGBRFClassifier()
xgb = XGBClassifier()

def model_score(model):
    score=[]
    model_name = type(model).__name__
    a = model.fit(X_train,y_train)
    a.score(X_test,y_test)
    y_pred = a.predict(X_test)
    acc = accuracy_score(y_test,y_pred)  
    return model_name,acc  



models = [xgbrf,xgb]
score =[]
for name in models:
    a,b = model_score(name)
    score.append([a,round(b*100,2)])
df = pd.DataFrame(data=score,columns=['name','accuracy_score'])
c = df[df.accuracy_score==df.accuracy_score.max()]


print (f'the best model is {c.name.values[0]} and its score is {df.accuracy_score.max()}')

```
